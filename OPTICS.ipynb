```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import OPTICS
clusmodel=OPTICS(min_samples=7).fit(X)
label=clusmodel.fit_predict(X)
```

```python
label
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=label
```

```python
supervised
```
